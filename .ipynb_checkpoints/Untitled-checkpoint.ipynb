```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# Load the extracted features and surface roughness values
#features_df = pd.read_csv('path_to_your_data/Extracted_Features.csv')
#y = features_df['Ra']
#X = features_df.drop(columns=['Ra'])



# Load the surface roughness values
surface_roughness = pd.read_csv('Machining quality.csv')['Ra']

# Combine features and target variable
X_selected['Ra'] = surface_roughness

y = X_selected['Ra']
X = X_selected.drop(columns=['Ra'])


# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform Sequential Feature Selection (SFS) using SVR
sfs = SequentialFeatureSelector(SVR(kernel='linear'), n_features_to_select=5, direction='forward')
sfs.fit(X_scaled, y)
selected_features = sfs.get_support(indices=True)
X_selected = X.iloc[:, selected_features]

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Defining the ANN model
ann = MLPRegressor(hidden_layer_sizes=(5, 5), max_iter=10000, random_state=42)
ann.fit(X_train, y_train)

# Predicting and evaluating
y_train_pred = ann.predict(X_train)
y_test_pred = ann.predict(X_test)

mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

print(f'Training MSE: {mse_train}')
print(f'Testing MSE: {mse_test}')
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# Function to extract statistical features
def extract_statistical_features(signal):
    features = {
        'mean': np.mean(signal),
        'minimum': np.min(signal),
        'maximum': np.max(signal),
        'kurtosis': pd.Series(signal).kurtosis(),
        'skewness': pd.Series(signal).skew(),
        'sum': np.sum(signal),
        'variance': np.var(signal),
        'range': np.ptp(signal),
        'rms': np.sqrt(np.mean(signal**2)),
        'srm': np.sqrt(np.mean(np.sqrt(np.abs(signal)))),
        'crest_factor': np.max(signal) / np.sqrt(np.mean(signal**2)),
        'latitude_factor': np.max(signal) / np.mean(np.abs(signal)),
        'impulse_factor': np.max(signal) / np.mean(signal)
    }
    return features

# Combine data from all workpieces
features_list = []
for i in range(1, 9):
    file_path = f'Workpiece#{i}.csv'
    data = pd.read_csv(file_path)
    # Assuming each file contains multiple samples
    for j in range(len(data)):
        features_x = extract_statistical_features(data['Rawdata_X'])
        features_y = extract_statistical_features(data['Rawdata_Y'])
        features_z = extract_statistical_features(data['Rawdata_Z'])
        combined_features = {f'{k}_x': v for k, v in features_x.items()}
        combined_features.update({f'{k}_y': v for k, v in features_y.items()})
        combined_features.update({f'{k}_z': v for k, v in features_z.items()})
        features_list.append(combined_features)

# Convert list of dictionaries to DataFrame
features_df = pd.DataFrame(features_list)
features_df
```

```python
# Load the surface roughness values
surface_roughness = pd.read_csv('Machining quality.csv')['Ra']

# Ensure the number of surface roughness values matches the number of samples
assert len(surface_roughness) == len(features_df), "Mismatch between number of samples and surface roughness values"

# Combine features and target variable
features_df['Ra'] = surface_roughness

# Standardize the features
scaler = StandardScaler()
X = features_df.drop(columns=['Ra'])
y = features_df['Ra']
X_scaled = scaler.fit_transform(X)

# Perform Sequential Feature Selection (SFS) using SVR
sfs = SequentialFeatureSelector(SVR(kernel='linear'), n_features_to_select=5, direction='forward')
sfs.fit(X_scaled, y)
selected_features = sfs.get_support(indices=True)
X_selected = X.iloc[:, selected_features]

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Defining the ANN model with specified parameters
ann = MLPRegressor(hidden_layer_sizes=(5, 5), max_iter=10000, random_state=42)
ann.fit(X_train, y_train)

# Predicting and evaluating
y_train_pred = ann.predict(X_train)
y_test_pred = ann.predict(X_test)

mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

# Perform cross-validation
cv_scores = cross_val_score(ann, X_scaled[:, selected_features], y, cv=5, scoring='neg_mean_squared_error')
cv_mse = -cv_scores.mean()

print(f'Training MSE: {mse_train}')
print(f'Testing MSE: {mse_test}')
print(f'Cross-Validation MSE: {cv_mse}')
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# Function to extract statistical features for a DataFrame
def extract_statistical_features(df):
    features = {
        'mean': df.mean(axis=0),
        'minimum': df.min(axis=0),
        'maximum': df.max(axis=0),
        'kurtosis': df.kurtosis(axis=0),
        'skewness': df.skew(axis=0),
        'sum': df.sum(axis=0),
        'variance': df.var(axis=0),
        'range': df.apply(np.ptp, axis=0),
        'rms': np.sqrt((df ** 2).mean(axis=0)),
        'srm': np.sqrt(np.sqrt(np.abs(df)).mean(axis=0)),
        'crest_factor': df.max(axis=0) / np.sqrt((df ** 2).mean(axis=0)),
        'latitude_factor': df.max(axis=0) / df.abs().mean(axis=0),
        'impulse_factor': df.max(axis=0) / df.mean(axis=0)
    }
    return pd.DataFrame(features).T

# Combine data from all workpieces
features_list = []
for i in range(1, 9):
    file_path = f'Workpiece#{i}.csv'
    data = pd.read_csv(file_path)
    data.columns = ['Sample_No', 'Rawdata_X', 'Rawdata_Y', 'Rawdata_Z']
    
    # Extract features for each direction
    features_x = extract_statistical_features(data[['Rawdata_X']])
    features_y = extract_statistical_features(data[['Rawdata_Y']])
    features_z = extract_statistical_features(data[['Rawdata_Z']])
    
    # Combine features into a single DataFrame
    combined_features = pd.concat([features_x, features_y, features_z], axis=1)
    features_list.append(combined_features)

# Convert list of DataFrames to a single DataFrame
features_df = pd.concat(features_list, ignore_index=True)
features_df
```

```python
import numpy as np
import pandas as pd

def extract_features(signal):
    features = {
        'mean': np.mean(signal),
        'minimum': np.min(signal),
        'maximum': np.max(signal),
        'kurtosis': pd.Series(signal).kurtosis(),
        'skewness': pd.Series(signal).skew(),
        'sum': np.sum(signal),
        'variance': np.var(signal),
        'range': np.ptp(signal),
        'rms': np.sqrt(np.mean(signal**2)),
        'srm': np.sqrt(np.mean(np.sqrt(np.abs(signal)))),
        'crest_factor': np.max(signal) / np.sqrt(np.mean(signal**2)),
        'latitude_factor': np.max(signal) / np.mean(np.sqrt(np.abs(signal))),
        'impulse_factor': np.max(signal) / np.mean(signal)
    }
    return features

# Load the vibration data from each workpiece file
workpiece_files = [
    'Workpiece#1.csv', 'Workpiece#2.csv', 'Workpiece#3.csv', 'Workpiece#4.csv',
    'Workpiece#5.csv', 'Workpiece#6.csv', 'Workpiece#7.csv', 'Workpiece#8.csv'
]

all_features = []

for file in workpiece_files:
    df = pd.read_csv(f'{file}')
    # Assuming the vibration data is in columns 'Rawdata_X', 'Rawdata_Y', 'Rawdata_Z'
    for axis in ['Rawdata_X', 'Rawdata_Y', 'Rawdata_Z']:
        signal = df[axis].values
        num_samples = 40
        sample_length = 2048
        
        # Ensure the signal can be divided into num_samples of sample_length
        if len(signal) < num_samples * sample_length:
            continue
        
        # Divide the signal into 40 samples
        for i in range(num_samples):
            start_idx = i * sample_length
            end_idx = start_idx + sample_length
            sample = signal[start_idx:end_idx]
            
            features = extract_features(sample)
            features['axis'] = axis
            features['workpiece'] = file
            all_features.append(features)

features_df = pd.DataFrame(all_features)

# Save the extracted features
#features_df.to_csv('vibration_features.csv', index=False)
features_df
```
